# Bias Variance decomposition analysis

```python
import sys; sys.path.append('../'); sys.path.append('../Preprocess')
import numpy as np
import pandas as pd

# use sklearn metrics, single function
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

import Preprocess.preprocessor as preprocessor
from imblearn.over_sampling import SMOTE

import xgboost as xgb
from lightgbm.sklearn import LGBMClassifier
from catboost import CatBoostClassifier

import mlxtend.evaluate.bias_variance_decomp as bias_variance_decomp
```

```python
# read data
df = preprocessor.ReadData(pth='../Dataset/body_level_classification_train.csv', label='Body_Level')
# aggregate columns, BMI index, ...
df = preprocessor.Aggregate(df, discretize=False)
# encode while preserving ordinal feature of data!
df = preprocessor.LabelOrdinalEncode(df)
#
df.head()
```

```python
X_train, X_test, y_train, y_test = preprocessor.Split(df, test_size=0.2, random_state=42)
print(y_train.value_counts())
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(random_state=123)

X_train = X_train.values
y_train = y_train.values
X_test = X_test.values
y_test = y_test.values

avg_expected_loss, avg_bias, avg_var = bias_variance_decomp(
        tree, X_train, y_train, X_test, y_test, 
        loss='0-1_loss',
        random_seed=123)

print('Average expected loss: %.3f' % avg_expected_loss)
print('Average bias: %.3f' % avg_bias)
print('Average variance: %.3f' % avg_var)
```
